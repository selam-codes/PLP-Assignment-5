```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('customers-100.csv')


```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.cleaned = df.dropna()
```

```python
df.cleaned.describe()
```

```python
df.cleaned.groupby('Country').mean(numeric_only=True)
```

```python
if 'Customer Id' in df.cleaned.columns and 'City' in df.cleaned.columns:
    plt.figure(figsize=(8,5))
    plt.plot(df.cleaned['Customer Id'], df.cleaned['City'])
    plt.title("Line Chart: City Over CustomerID")
    plt.xlabel("CustomerID")
    plt.ylabel("Age")
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
if 'Customer Id' in df.cleaned.columns and 'City' in df.cleaned.columns:
    plt.figure(figsize=(8,5))
    sns.barplot(x='Customer Id', y='City', data=df.cleaned)
    plt.title("Bar Chart: City Over CustomerID")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    
```

```python
if 'City' in df.cleaned.columns:
    plt.figure(figsize=(8,5))
    plt.hist(df.cleaned['City'], bins=15, color='skyblue', edgecolor='black')
    plt.title("Histogram: City Distribution")
    plt.xlabel("City")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

```python
if 'City' in df.cleaned.columns and 'Subscription Date' in df.cleaned.columns:
    plt.figure(figsize=(8,5))
    sns.scatterplot(x='City', y='Subscription Date', data=df.cleaned)
    plt.title("Scatter Plot: City Vs Subscription Date")
    plt.xlabel("City")
    plt.ylabel("Subscription Date")
    plt.tight_layout()
    plt.show()
```
